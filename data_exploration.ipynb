```python
import pandas as pd

# Replace 'file.xlsx' with the path to your Excel file
df = pd.read_excel("data_cts_violent_and_sexual_crime.xlsx")

df["Iso3_code"] = df["Iso3_code"].replace({
    "GBR_E_W": "ENG",
    "GBR_NI": "NIR",
    "GBR_S": "SCO"
})
df.rename(columns={"Iso3_code": "Code"}, inplace=True)
```

```python
category_of_crime = "Serious assault"
unit_of_measurement = "Rate per 100,000 population"
year = 2022

df_assault = df[(df["Category"] == category_of_crime) & (df["Unit of measurement"] == unit_of_measurement) & (df["Year"] == year)]
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Create the bar chart
plt.figure(figsize=(24,8))
plt.bar(df_assault["Code"], df_assault["VALUE"])

# Add labels and title
plt.xlabel("Code", labelpad=40)
plt.ylabel(unit_of_measurement)
plt.title(f"Bar Chart of {category_of_crime} ({unit_of_measurement}) for every country")

# Get the x-tick positions and labels
x_ticks = np.arange(len(df_assault["Code"]))
x_labels = df_assault["Code"]

# Alternate the label positions
for i, label in enumerate(x_labels):
    if i % 2 == 0:
        plt.text(x_ticks[i], -30, label, ha='center', rotation=0, va='top', fontsize=9)
    else:
        plt.text(x_ticks[i], -80, label, ha='center', rotation=0, va='top', fontsize=9)

# Remove the default x-axis labels
plt.xticks([])

# Show the plot
plt.tight_layout()
plt.show()
```
